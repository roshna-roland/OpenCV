```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/dog.jpeg")
```

```python
g=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(g,cmap='gray')
```

```python
#canny edge detection
ed=cv2.Canny(g, threshold1=30,threshold2=100)
```

```python
plt.imshow(ed, cmap='gray')
plt.show()
```

```python
#harris corner detection
corner=cv2.cornerHarris(g,2,3,0.04)    #detect corner
corner=cv2.dilate(corner,None)         #dilate
img[corner>0.01 *corner.max()]=[0,0,255]
plt.imshow(corner,cmap='gray')
plt.show()
```

```python
edges=cv2.Canny(g,75,150)
lines=cv2.HoughLinesP(edges, 1,np.pi/180,30,maxLineGap=250)
for line in lines:
  x1,y1,x2,y2=line[0]
  cv2.line(img,(x1,y1),(x2,y2),(0,0,128),1)
plt.imshow(edges,cmap='gray')
plt.show()
```
